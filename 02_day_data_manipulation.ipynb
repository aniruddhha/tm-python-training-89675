```python
# Dimensions: Branches x Departments x Quarters

expenses = np.array([
    [  # Branch 1
        [15000, 15500, 14800, 16000],    # Sales Dept
        [8000, 8500, 8200, 9000],        # HR Dept
        [12000, 12300, 11900, 12500]     # Operations Dept
    ],
    [  # Branch 2
        [14200, 15000, 14700, 15800],    # Sales Dept
        [7800, 8100, 7900, 8500],        # HR Dept
        [11700, 12200, 11800, 12400]     # Operations Dept
    ]
])

print(f'Dimensions {expenses.ndim}')
print(f'Shape {expenses.shape}')
print(f'Size {expenses.size}')
```

## ***Case Study***

```python
# axis = dimensions, we have 3d system
# 0 - Branch, 1 - Department, 2 - Quarter

expenses = np.array([
     [  # Branch 1
        [15000, 15500, 14800, 16000, 16200],    # Sales Dept
        [8000, 8500, 8200, 9000, 9100],         # HR Dept
        [12000, 12300, 11900, 12500, 12600]     # Operations Dept
    ],
    [  # Branch 2
        [14200, 15000, 14700, 15800, 15900],    # Sales Dept
        [7800, 8100, 7900, 8500, 8800],         # HR Dept
        [11700, 12200, 11800, 12400, 12650]     # Operations Dept
    ],
    [  # Branch 3
        [15500, 15700, 15400, 16300, 16600],    # Sales Dept
        [8200, 8700, 8450, 9200, 9500],         # HR Dept
        [12500, 12800, 12450, 13000, 13300]     # Operations Dept
    ]
])
```

```python
# Average Expense (Overall)
print(np.mean(expenses))
```

```python
# Average Expense Per Branch
print(np.mean(expenses, axis=(1, 2)))
```

```python
# Average Expense Per Department (across all branches & quarters)
print(np.mean(expenses, axis=(0, 2)))
```

```python
# Average Expense Per Quarter (across all branches & departments)
print(np.mean(expenses, axis=(0, 1)))
```

```python
# Min and Max Expense
min_expense = np.min(expenses)
max_expense = np.max(expenses)
print("Minimum expense:", min_expense)
print("Maximum expense:", max_expense)
```

```python
# Total Expenses Per Branch
total_per_branch = np.sum(expenses, axis=(1,2))
print("Total expenses per branch:", total_per_branch)
```

```python
transactions = pd.DataFrame({
    'Date': ['2024-04-01', '2024-04-01'],
    'AccountCode': [101, 102],
    'Amount': [2000, 1500]
})

accounts = pd.DataFrame({
    'Date': ['2024-04-01', '2024-04-01'],
    'AccountCode': [101, 102],
    'AccountName': ['Cash', 'Office Supply']
})

transactions['Key'] = transactions['Date'].astype(str) + "_" + transactions['AccountCode'].astype(str)
accounts['Key'] = accounts['Date'].astype(str) + "_" + accounts['AccountCode'].astype(str)

merged = pd.merge(transactions, accounts[['Key', 'AccountName']], on='Key', how='left')

print(merged)
```
